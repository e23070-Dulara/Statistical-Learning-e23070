```python
import math

# Probabilities
p = {
    ("omega1", 0, 0): 0.50,
    ("omega2", 1, 0): 0.10,
    ("omega3", 0, 1): 0.20,
    ("omega4", 1, 1): 0.20
}

# Marginals
P_Y0 = 0.50 + 0.10
P_Y1 = 0.20 + 0.20

P_X1 = 0.10 + 0.20
P_X0 = 0.50 + 0.20

# Conditional expectations
EX_Y0 = 0.10 / 0.60
EX_Y1 = 0.20 / 0.40

print("E[X|Y=0] =", EX_Y0)
print("E[X|Y=1] =", EX_Y1)

# Conditional PMF
print("\nConditional PMF:")
print("P(X=0|Y=0) =", 0.50 / 0.60)
print("P(X=1|Y=0) =", 0.10 / 0.60)

print("P(X=0|Y=1) =", 0.20 / 0.40)
print("P(X=1|Y=1) =", 0.20 / 0.40)

# MSE using conditional expectation
var0 = EX_Y0 * (1 - EX_Y0)
var1 = EX_Y1 * (1 - EX_Y1)

MSE_conditional = P_Y0 * var0 + P_Y1 * var1

print("\nConditional MSE =", MSE_conditional)

# Global estimator
EX = P_X1
MSE_global = EX * (1 - EX)

print("Global-average MSE =", MSE_global)

# Entropy function
def entropy(probs):
    h = 0
    for prob in probs:
        if prob > 0:
            h -= prob * math.log2(prob)
    return h

# Entropies
HX = entropy([P_X0, P_X1])

HX_Y0 = entropy([0.50/0.60, 0.10/0.60])
HX_Y1 = entropy([0.20/0.40, 0.20/0.40])

HXY = P_Y0 * HX_Y0 + P_Y1 * HX_Y1

# Mutual information
IXY = HX - HXY

percentage = (IXY / HX) * 100

print("\nH(X) =", HX)
print("H(X|Y) =", HXY)
print("I(X;Y) =", IXY)

print("\nPercentage uncertainty removed =", percentage, "%")
```

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

```python
import numpy as np

# Example dimensions
n = 2
m = 1

# Mean vectors
mu_x = np.array([1, 2])
mu_y = np.array([3])

# Covariance blocks
Sigma_xx = np.array([
    [2, 0.5],
    [0.5, 1]
])

Sigma_xy = np.array([
    [0.3],
    [0.2]
])

Sigma_yx = Sigma_xy.T

Sigma_yy = np.array([[1.5]])

# Observed value of Y
y = np.array([4])

# Conditional mean
Sigma_yy_inv = np.linalg.inv(Sigma_yy)

mu_star = mu_x + Sigma_xy @ Sigma_yy_inv @ (y - mu_y)

# Conditional covariance
Sigma_cond = Sigma_xx - Sigma_xy @ Sigma_yy_inv @ Sigma_yx

print("Conditional Mean:")
print(mu_star)

print("\nConditional Covariance:")
print(Sigma_cond)

# Example x point
x = np.array([1.5, 2.5])

# Multivariate Gaussian conditional density
k = n

det_cov = np.linalg.det(Sigma_cond)
inv_cov = np.linalg.inv(Sigma_cond)

diff = x - mu_star

density = (
    1
    / (((2 * np.pi) ** (k / 2)) * np.sqrt(det_cov))
    * np.exp(-0.5 * diff.T @ inv_cov @ diff)
)

print("\nConditional Density f_X|Y(x|y):")
print(density)
```

# Q3) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**
1.  **Hidden Temperature ($X$):** The standardized temperature deviation follows a standard normal distribution:
    $$X \sim \mathscr{N}(0, 1)$$
2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

---

**Answer the followings:**

**1. Joint Distribution**
Identify the distribution of the random varaible $(X, Y)$.
* Show that $Y \sim \mathscr{N}(0, 5)$.
* Find the covariance $\text{Cov}(X, Y)$.

**2. Information Geometry**
Explain why $X$ is **not** $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?

**3. Conditional Expectation (The Best Estimate)**
Compute the conditional expectation $E[X \mid Y]$.
* Using the property for joint normals: $E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$.
* Express this as a "signal-to-total-variance" ratio.

**4. Verification of the Residual**
Let $\widehat{X} = E[X \mid Y]$. Define the estimation error (residual) as $Z = X - \widehat{X}$.
* Prove that $E[Z] = 0$.
* Verify that the error $Z$ is uncorrelated with the observation $Y$ (the Orthogonality Principle).

Why Y is a Poor Estimator

If Y=10, it is unlikely that all 10 units came from temperature because the noise variance is large.

Part of the observation is probably caused by mechanical load Q.

Thus using Y directly overestimates extreme temperatures.

Shrinkage Toward the Mean

The estimator:

E[X∣Y]=
5
1
	​

Y

shrinks observations toward zero because:

only 1/5 of the total observed variance comes from the true temperature signal,
the remaining 4/5 is noise.

Hence the optimal estimator discounts noisy observations.

```python
import numpy as np
import math

# Given variances
var_X = 1
var_Q = 4

# Y = X + Q
var_Y = var_X + var_Q

# Covariance
cov_XY = var_X

print("Var(Y) =", var_Y)
print("Cov(X,Y) =", cov_XY)

# Conditional expectation coefficient
coef = cov_XY / var_Y

print("\nE[X|Y=y] = ", coef, "* y")

# Conditional variance
var_cond = var_X - (cov_XY**2) / var_Y

print("Conditional Variance Var(X|Y) =", var_cond)

# Differential entropy function
def gaussian_entropy(variance):
    return 0.5 * math.log(2 * math.pi * math.e * variance)

# Entropies
h_X = gaussian_entropy(var_X)
h_X_given_Y = gaussian_entropy(var_cond)

print("\nh(X) =", h_X)
print("h(X|Y) =", h_X_given_Y)

# Mutual Information
I_XY = h_X - h_X_given_Y

print("I(X;Y) =", I_XY)

# Percentage uncertainty removed
percentage = (I_XY / h_X) * 100

print("Percentage uncertainty removed =", percentage, "%")

# Residual orthogonality
# Z = X - (1/5)Y

cov_ZY = cov_XY - coef * var_Y

print("\nCov(Z,Y) =", cov_ZY)

# High noise example
var_Q2 = 100
var_Y2 = var_X + var_Q2

I_high_noise = 0.5 * math.log(var_Y2 / var_Q2)

print("\nWith noise variance 100:")
print("Mutual Information =", I_high_noise)
```
