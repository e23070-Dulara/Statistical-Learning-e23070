```python
# Beta Distribution Model for Defective Proportion

import math
from scipy.special import beta
from scipy.integrate import quad

# Parameters
alpha = 2
beta_param = 18

# ------------------------------------------------
# 1. Probability Space
# ------------------------------------------------
Omega = "[0,1]"
F = "Borel sigma-algebra on [0,1]"

print("Sample Space Ω =", Omega)
print("Sigma-algebra F =", F)

# ------------------------------------------------
# 2. Beta Function
# ------------------------------------------------
B = beta(alpha, beta_param)

print("\nBeta Function B(2,18) =", B)

# ------------------------------------------------
# 3. Probability Density Function
# ------------------------------------------------
def f(x):
    if 0 < x < 1:
        return (1 / B) * (x**(alpha - 1)) * ((1 - x)**(beta_param - 1))
    else:
        return 0

print("\nPDF:")
print("f(x) = (1/B(2,18)) * x * (1-x)^17 , for 0 < x < 1")

# ------------------------------------------------
# 4. Verify PDF
# ------------------------------------------------
integral, _ = quad(f, 0, 1)

print("\nVerification of PDF:")
print("Integral of f(x) over [0,1] =", integral)

# ------------------------------------------------
# 5. Expectation E[X]
# ------------------------------------------------
EX = alpha / (alpha + beta_param)

print("\nExpectation E[X] =", EX)

# ------------------------------------------------
# 6. Variance Var(X)
# ------------------------------------------------
VarX = (alpha * beta_param) / (
    ((alpha + beta_param)**2) * (alpha + beta_param + 1)
)

print("Variance Var(X) =", VarX)

# ------------------------------------------------
# 7. Probability P(X <= 0.10)
# ------------------------------------------------
P1, _ = quad(f, 0, 0.10)

print("\nP(X <= 0.10) =", P1)

# ------------------------------------------------
# 8. Probability P(0.05 <= X <= 0.15)
# ------------------------------------------------
P2, _ = quad(f, 0.05, 0.15)

print("P(0.05 <= X <= 0.15) =", P2)

# ------------------------------------------------
# 9. Interpretation
# ------------------------------------------------
print("\nInterpretation:")
print(f"Average defective proportion = {EX:.2f} (10%)")
print(f"Variance = {VarX:.5f}, indicating variability in quality.")
```

```python
# Triangular Distribution Example
# Backup Generator Installation Time

import math
from scipy.integrate import quad

# Parameters
a = 4   # minimum
b = 10  # maximum
c = 6   # most likely

# ------------------------------------------------
# 1. Support
# ------------------------------------------------
print("Support of X: [4, 10]")

# ------------------------------------------------
# 2. Probability Density Function
# ------------------------------------------------
def f(x):
    if a <= x <= c:
        return 2 * (x - a) / ((b - a) * (c - a))
    elif c < x <= b:
        return 2 * (b - x) / ((b - a) * (b - c))
    else:
        return 0

print("\nPDF defined successfully.")

# ------------------------------------------------
# 3. Verify PDF integrates to 1
# ------------------------------------------------
integral, _ = quad(f, a, b)

print("\nIntegral of PDF over support =", integral)

# ------------------------------------------------
# 4. Expectation E[X]
# ------------------------------------------------
EX = (a + b + c) / 3

print("\nExpectation E[X] =", EX)

# ------------------------------------------------
# 5. Variance Var(X)
# ------------------------------------------------
VarX = (
    a**2 + b**2 + c**2 - a*b - a*c - b*c
) / 18

print("Variance Var(X) =", VarX)

# ------------------------------------------------
# 6. Probability P(X <= 5)
# ------------------------------------------------
P1, _ = quad(f, a, 5)

print("\nP(X <= 5) =", P1)

# ------------------------------------------------
# 7. Probability P(5 <= X <= 8)
# ------------------------------------------------
P2, _ = quad(f, 5, 8)

print("P(5 <= X <= 8) =", P2)

# ------------------------------------------------
# 8. Interpretation
# ------------------------------------------------
print("\nInterpretation:")
print(f"Minimum possible time = {a} hours")
print(f"Maximum possible time = {b} hours")
print(f"Most likely time = {c} hours")

print(f"\nAverage installation time ≈ {EX:.2f} hours")
print(f"Variance ≈ {VarX:.2f}, showing variability in completion time.")
```

9. Why PERT May Be More Appropriate than the Triangular Distribution

The PERT distribution is often preferred because:

It produces a smoother probability curve.
It gives more weight to the most likely value c.
It avoids the sharp corners of the Triangular distribution.
It better reflects realistic uncertainty in project durations.

Compared with the Triangular model, the PERT distribution usually provides more stable and less extreme estimates for scheduling and risk analysis.

10. Interpretation in Project Planning
a=8: shortest possible completion time under ideal conditions.
b=20: longest possible completion time under difficult conditions.
c=11: most likely completion time estimated by experts.
E[X]=12: expected average project duration considering uncertainty.
σ≈2.14: typical variability around the expected duration.

```python
# PERT Distribution Example
# Software Integration Task Duration

import math
from scipy.special import beta

# ------------------------------------------------
# Parameters
# ------------------------------------------------
a = 8    # minimum time
b = 20   # maximum time
c = 11   # most likely time

# ------------------------------------------------
# 1. Shape Parameters
# ------------------------------------------------
alpha = 1 + 4 * ((c - a) / (b - a))
beta_param = 1 + 4 * ((b - c) / (b - a))

print("Shape Parameters:")
print("alpha =", alpha)
print("beta =", beta_param)

# ------------------------------------------------
# 2. Support
# ------------------------------------------------
print("\nSupport of X:")
print(f"X belongs to [{a}, {b}]")

# ------------------------------------------------
# 3. Probability Density Function
# ------------------------------------------------
B = beta(alpha, beta_param)

def f(x):
    if a <= x <= b:
        numerator = ((x - a)**(alpha - 1)) * ((b - x)**(beta_param - 1))
        denominator = B * ((b - a)**(alpha + beta_param - 1))
        return numerator / denominator
    else:
        return 0

print("\nPDF successfully defined.")

# ------------------------------------------------
# 4. Expectation
# ------------------------------------------------
EX = (a + 4*c + b) / 6

print("\nExpectation E[X] =", EX)

# ------------------------------------------------
# 5. Variance
# ------------------------------------------------
VarX = ((EX - a) * (b - EX)) / 7

print("Variance Var(X) =", VarX)

# ------------------------------------------------
# 6. Standard Deviation
# ------------------------------------------------
sigma = math.sqrt(VarX)

print("Standard Deviation sigma =", sigma)

# ------------------------------------------------
# 7. Interpretation
# ------------------------------------------------
print("\nInterpretation:")
print(f"Minimum possible duration = {a} days")
print(f"Maximum possible duration = {b} days")
print(f"Most likely duration = {c} days")
print(f"Expected project duration = {EX:.2f} days")
print(f"Standard deviation = {sigma:.2f} days")

# ------------------------------------------------
# 8. Why PERT is Useful
# ------------------------------------------------
print("\nPERT distribution gives smoother and more realistic")
print("project duration estimates than the Triangular distribution.")
```
