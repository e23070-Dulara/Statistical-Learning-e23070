````python
# Interpretation of Sensor_4 with Uniqueness ≈ 98%

uniqueness_sensor4 = 0.98

if uniqueness_sensor4 > 0.95:
    print("Sensor_4 is dominated by unique/random variance.")
    print("Very little of its behavior is explained by common latent factors.")

# Why PCA can be misled

# PCA decomposes TOTAL variance:
# Total Variance = Common Variance + Unique Variance + Measurement Noise

# Example conceptually:
common_variance = 0.02   # 2%
unique_variance = 0.98   # 98%

total_variance = common_variance + unique_variance

# PCA only sees:
print("PCA maximizes:", total_variance)

# FA separates:
print("Common variance:", common_variance)
print("Unique variance:", unique_variance)

# Consequence:
# A high-noise sensor contributes a large amount of variance.
# PCA may allocate principal component weight to that sensor,
# increasing the leading eigenvalues even though the variance
# does not represent meaningful plant behavior.

# Risk in anomaly detection:
# Model may learn noise patterns rather than true system dynamics.

if unique_variance > common_variance:
    print("Warning: PCA-based anomaly detector may generate")
    print("false alarms and miss real process faults.")
````

A uniqueness value above 98% means:

Almost all variation in Sensor_4 is independent of the underlying system behavior.
Sensor_4 contributes very little information about the shared process dynamics.
The sensor is likely dominated by:
Measurement noise,
Local disturbances,

Sensor-specific effects,

2.

PCA searches for directions that maximize total variance.

Mathematically, PCA does not distinguish between:

Useful process variance,
Shared system variance,
Random measurement noise.

Since Sensor_4 contains extremely large variance (98% unique variance), PCA interprets that variance as important simply because it is numerically large.

As a result:

Sensor_4 contributes heavily to the covariance matrix.
The covariance matrix produces larger eigenvalues.

PC1 and PC2 appear to explain a large amount of variance, but part of that variance is meaningless sensor-specific noise.

Thus, PCA can mistake "large variance" for "important variance."

Factor Analysis avoids this problem because it explicitly separates:

Total Variance=Common Variance+Unique Variance

and recognizes that most of Sensor_4's variance is not shared with the rest of the system.

3.

If an engineer builds an anomaly detector using only PCA:

False Alarms

### Question 3: Determining Subspace Truncation ($k$) using $T^2$ and $Q$ Profiles

* **Context:** Look at the right side of the PCA dashboard, showing **Mean Hotelling's $T^2$** and **Mean $Q$ Statistic (SPE)** plotted against the truncation cutoff $k$.
* **Core Problem:**

2. Why does a sharp drop followed by a distinct "flat elbow" in the $Q$ statistic identify **$k=2$** as the true hidden physical dimensionality of this asset?
3. If you incorrectly choose a truncation cutoff of $k=3$ for your monitoring system, what type of data are you forcing into your "clean" principal subspace?

````python
# PCA Subspace Selection using Q Statistic (Squared Prediction Error)

# Example behavior observed in the dashboard

mean_Q = {
    1: 12.8,   # Large residual error
    2: 1.1,    # Sharp drop
    3: 0.9     # Very small improvement
}

# Interpretation

for k in range(1, 4):
    print(f"k={k}, Mean Q={mean_Q[k]}")

# Change in Q statistic

drop_1_to_2 = mean_Q[1] - mean_Q[2]
drop_2_to_3 = mean_Q[2] - mean_Q[3]

print("Drop from k=1 to k=2:", drop_1_to_2)
print("Drop from k=2 to k=3:", drop_2_to_3)

# Decision rule

if drop_1_to_2 >> drop_2_to_3:
    print("Elbow detected at k=2")
    print("Choose k=2 as the process dimensionality")

# Conceptual explanation:
#
# Total Data Space
# = Principal Subspace + Residual Subspace
#
# Q Statistic measures:
# Residual reconstruction error
#
# Large drop:
# Important process structure captured.
#
# Flat region:
# Remaining variance is mostly noise.
#
# Choosing k=3:
# Noise begins entering the principal subspace.

chosen_k = 3

if chosen_k > 2:
    print("Warning:")
    print("Sensor noise and minor random variation")
    print("are being treated as meaningful process behavior.")
````

1. **The PCA Strategy:** Tracking the main principal subspace using Hotelling's $T^2$ alongside a residual noise floor monitor using the $Q$ statistic.
2. **The FA Strategy:** Monitoring the rotated Latent Factor Scores directly.

In a real-time predictive maintenance system, both PCA and Factor Analysis can be used to monitor asset health, but they do so in fundamentally different ways.

1. PCA Strategy: Monitoring T² and Q Statistics

The PCA approach divides the data into two regions:

Principal Subspace → monitored using Hotelling's T² statistic.
Residual Subspace → monitored using the Q Statistic (Squared Prediction Error).

The T² statistic measures how unusual an observation is within the retained principal component space. It is useful for detecting shifts in the dominant operating modes of the asset.

The Q statistic measures the amount of unexplained variation remaining after projection onto the principal subspace. It is effective at detecting abnormal behavior that does not follow the learned process structure.

This dual-monitoring framework provides two complementary views:

T² detects deviations in modeled system behavior.
Q detects unexpected residual disturbances and sensor anomalies.

However, PCA does not explicitly distinguish between common process variation and sensor-specific variation. All variance contributes to the covariance structure used to build the model.

2. FA Strategy: Monitoring Rotated Latent Factor Scores

Factor Analysis focuses only on the common variance shared among sensors.

The rotated latent factors represent interpretable physical subsystems or process mechanisms. Monitoring factor scores therefore means directly monitoring the underlying system states rather than the raw sensor measurements.

Because FA explicitly models:

Total Variance=Common Variance+Unique Variance

it can separate process-related information from sensor-specific noise.

As a result:

Factor scores respond primarily to genuine system changes.
Sensor-specific disturbances have limited influence on the latent factors.
The model is generally easier to interpret and diagnose.
Which Strategy Is More Robust to Sensor Calibration Loss or an Electrical Short?

The Factor Analysis strategy is generally more robust when a single sensor loses calibration or develops an electrical fault.

The key evidence comes from the Sensor Uniqueness Noise Floor (ϕ
2
) values.

ϕ

This means approximately 98% of its variance is unique to that sensor and is not part of the common process structure.

Factor Analysis recognizes this immediately:

Sensor_4 contributes very little information to the latent factors.
Changes in Sensor_4 have minimal impact on the factor scores.
The common process model remains largely unaffected.

In effect, the faulty sensor is automatically isolated from the shared system representation.

Why PCA Is Less Robust

PCA does not separate common and unique variance.

If Sensor_4 develops:

Calibration drift,
Electrical noise,
Signal spikes,
A short circuit,

the additional variance can influence the covariance matrix and affect principal component projections.

Even if the fault eventually appears in the Q statistic, the sensor's variance may still distort the PCA representation because PCA treats all variance as potentially important.

As a result:

Principal components may become partially influenced by the faulty sensor.
False alarms may occur.
Model interpretation becomes more difficult.
